```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Load data
dataset = pd.read_csv('./Weight-Height Polynomial Dataset.csv')
print(dataset)
X = dataset.iloc[:, 0:1].values
y = dataset.iloc[:, 1].values
```

```python
# plot data point
plt.scatter(X, y, color = 'blue')
plt.xlabel('Weight')
plt.ylabel('Height')
plt.show()
```

```python
# Train Linear model
from sklearn.linear_model import LinearRegression
linear_reg_model = LinearRegression()
linear_reg_model.fit(X, y)
```

```python
# Visualize Linear regression
plt.scatter(X, y, color = 'blue')
plt.plot(X, linear_reg_model.predict(X), color = 'green')
plt.title('Linear Regression')
plt.xlabel('Weight')
plt.ylabel('Height')
plt.show()
```

```python
from sklearn.preprocessing import PolynomialFeatures
polynomial_ft = PolynomialFeatures(degree = 7)
X_polynomial = polynomial_ft.fit_transform(X)
```

```python
# Train polynomial regression model
polynomial_reg_model = LinearRegression()
polynomial_reg_model.fit(X_polynomial, y)
```

```python
# Visualize polynomial regression
plt.scatter(X, y, color = 'blue')
plt.plot(X, polynomial_reg_model.predict(polynomial_ft.fit_transform(X)), color = 'green')
plt.title('Polynomial Regression')
plt.xlabel('Weight')
plt.ylabel('Height')
plt.show()
```

```python
# Predict salary using linear regression
print(linear_reg_model.predict([[6.8]]))

# Predict salary using polynomial regression
print(polynomial_reg_model.predict(polynomial_ft.fit_transform([[6.8]])))
```
